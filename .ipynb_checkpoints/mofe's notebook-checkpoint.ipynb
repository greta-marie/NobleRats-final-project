```python
# import pandas
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

# load data
smart_trash = pd.read_csv("smart_trash.csv") 
smart_trash.head(10)

#, index_col="neighborhood"
```

```python
more_than_1 = lambda x, threshold: x['council_district'].count() > 1

least_councils = smart_trash.groupby("council_district").filter(more_than_1, threshold = 1)

least_councils.groupby("council_district").count()
```
